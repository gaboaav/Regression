## Select Test and Train Data

```python
King_Country_Test=pd.read_csv('kc_house_test_data.csv',sep=',',dtype=dtype_dic)
King_Country_Train=pd.read_csv('kc_house_train_data.csv',sep=',',dtype=dtype_dic)
```

## Linear Regression

```python
Train_Input=King_Country_Train['sqft_living']
Train_Output=King_Country_Train['price']
```

```python
def simple_linear_regression(IN,OUT):
    N=len(IN)
    Zx=IN.sum()
    Zy=OUT.sum()
    ZxZy=Zx*Zy
    Zxy=(IN*OUT).sum()
    Zxx=(IN*IN).sum()
    ZxZx=Zx*Zx
    slope=(Zxy-(ZxZy)/N)/(Zxx-(ZxZx)/N)
    intercept=(Zy/N)-slope*(Zx/N)
    return (intercept,slope)
```

```python
intercept,slope=simple_linear_regression(Train_Input,Train_Output)
print("The intercept is equal to: ",intercept)
print('The slope is equal to: ',slope)
```

## Predictions

```python
def predict(IN,Slope,Intercept):
    prediction=Intercept+Slope*IN
    return prediction

predict(2650, slope, intercept)
```

## Residual Sum of Squares

```python
def RSS(IN,OUT,Slope,Intercept):
    Yp=Intercept+Slope*IN
    RSS=(pow(OUT-Yp,2)).sum()
    return RSS
RSS(Train_Input,Train_Output,slope,intercept)
```

## Inverse Predictions

```python
def inv_pred(OUT,Intercept,Slope):
    IN=(OUT-Intercept)/Slope
    return IN
inv_pred(800000,intercept,slope)
```

## New Simple Regression

```python
Train_Input_N=King_Country_Train['bedrooms']
intercept2,slope2=simple_linear_regression(Train_Input_N,Train_Output)
print("The intercept is equal to: ",intercept2)
print('The slope is equal to: ',slope2)
```

## Trying With Test Data

```python
Test_Input=King_Country_Test['sqft_living']
Test_Input_N=King_Country_Test['bedrooms']
Test_Output=King_Country_Test['price']
```

```python
Rss1=RSS(Test_Input,Test_Output,slope,intercept)
Rss2=RSS(Test_Input_N,Test_Output,slope2,intercept2)
print('1.- {}  ;  2.- {}'.format(Rss1,Rss2))
```
