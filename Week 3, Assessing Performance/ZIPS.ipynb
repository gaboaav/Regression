# Open ZipFiles

```python
from zipfile import ZipFile

HD='kc_house_data.csv.zip'
HS1D='kc_house_set_1_data.csv.zip'
HS2D='kc_house_set_2_data.csv.zip'
HS3D='kc_house_set_3_data.csv.zip'
HS4D='kc_house_set_4_data.csv.zip'
HTeD='kc_house_test_data.csv.zip'
HTrD='kc_house_train_data.csv.zip'
HVD='kc_house_valid_data.csv.zip'

with ZipFile(HD,'r') as zip:
    print('HOUSE DATA\n')
    zip.printdir()
    zip.extractall()
    print('\nDone\n')

with ZipFile(HS1D,'r') as zip:
    print('HOUSE DATA SET 1\n')
    zip.printdir()
    zip.extractall()
    print('\nDone\n')
    
with ZipFile(HS2D,'r') as zip:
    print('HOUSE DATA SET 2\n')
    zip.printdir()
    zip.extractall()
    print('\nDone\n')
    
with ZipFile(HS3D,'r') as zip:
    print('HOUSE DATA SET 3\n')
    zip.printdir()
    zip.extractall()
    print('\nDone\n')
    
with ZipFile(HS4D,'r') as zip:
    print('HOUSE DATA SET 4\n')
    zip.printdir()
    zip.extractall()
    print('\nDone\n')
    
with ZipFile(HTeD,'r') as zip:
    print('HOUSE TEST DATA\n')
    zip.printdir()
    zip.extractall()
    print('\nDone\n')
    
with ZipFile(HTrD,'r') as zip:
    print('HOUSE TRAIN DATA\n')
    zip.printdir()
    zip.extractall()
    print('\nDone\n')
    
with ZipFile(HVD,'r') as zip:
    print('HOUSE VALID DATA\n')
    zip.printdir()
    zip.extractall()
    print('\nDone\n')
```
