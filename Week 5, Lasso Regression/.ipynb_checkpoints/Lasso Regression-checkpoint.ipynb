# Overfitting Demo

## Create a Dataset based on a true sinusoidal relationship

Let's look at a synthetic dataset consisting of 30 points drawn from the sinusoid y=sin(4x)

```python
import pandas as pd
import math as m
import random
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression

%matplotlib inline
```

Create random values for x in interval [0,1]

```python
random.seed(98103)
n=30
x=pd.array(sorted([random.random() for i in range(n)]))
x
```

Compute "y"

```python
def Compute(x):
    y=[]
    for i in x:
        j=m.sin(4*i)
        y.append(j)
    x=pd.array(y)
    return x

y=Compute(x)
y
```

Add random Gaussian noise to y

```python
random.seed(1)
e=pd.array([random.gauss(0,1/3) for i in range(n)])
y=y+e
```

## Put data into an DataFrame to manipulate later

```python
data=pd.DataFrame({'x1':x,'y':y})
data
```

## Create a function to plot the data

```python
def plot_data(data):
    plt.plot(data['x1'],data['y'],'k.')
    plt.xlabel('x')
    plt.ylabel('y')
    
plot_data(data)
```

## Define some useful polynomial regression functions

Define a function to create our features for a polynomial regression model of any degree:

```python
def poly_fts(data, deg):
    if 'y' in data.columns:
        data=data.drop(columns=['y'])
    copy=data.copy()
    for i in range(1,deg):
        copy['x'+str(i+1)]=copy['x'+str(i)]*copy['x1']
    return copy
```

Define a function to fit a polynomial linear regression model of degree "deg" to the data in "data"

```python
def poly_reg(data,deg):
    feats=poly_fts(data,deg)
    if deg==1:
        x=data.iloc[:,:-1].values
        y=data.iloc[:,1].values
        Model=LinearRegression()
        Model.fit(x,y)
        
    else:
        x=feats
        y=data['y']
        Model=LinearRegression()
        Model.fit(x,y)
    
    return Model
```

Define function to plot data and predictios made

```python
def PPP(data, model):
    plot_data(data)
    deg=len(model.coef_)
    
    x_pred=pd.DataFrame({'x1':[i/200 for i in range(200)]})
    y_pred=model.predict(poly_fts(x_pred,deg))
    
    plt.plot(x_pred['x1'],y_pred, 'g-', label='degree '+str(deg)+' fit')
    plt.legend(loc='upper left')
    plt.axis([0,1,-1.5,2])
```

Create a function that prints the polynomial coefficients

```python
def PCoeff(model):
    deg=len(model.coef_)
    w=list(model.coef_)
    print('Learned polynomial for degree '+str(deg)+':')
    w.reverse()
    w.append(model.intercept_)
    print(np.poly1d(w))
```

## Fit a 2nd Degree Polynomial

Fit our 2nd degree polynomial to the data generated above:

```python
model=poly_reg(data,2)
```

Inspect Parameters

```python
PCoeff(model)
```

Form and plot our preductions along a grid of x values

```python
PPP(data,model)
```

## Fit a 4th Degree Polynomial

```python
model=poly_reg(data,4)
PCoeff(model)
PPP(data,model)
```

## Fit a 16th Degree Polynomial

```python
model=poly_reg(data,16)
PCoeff(model)
PPP(data,model)
```

















# Ridge Regression

```python
from sklearn.linear_model import Ridge
```

Ridge regression aims to avoid overfitting by adding a cost to the RSS term of standard least squares that depends on the 2-norm of the coefficients ||w||. The result is penalizing fits with large coefficients. The strength of the penalty, and thus the fit vs model complexity balance is controled by a parameter lambda (here called "L2_penalty")

Define our function to solve the ridge objective for a polynomial regression model of any degree:

```python
def poly_ridge_reg(data,deg, alpha):
    feats=poly_fts(data,deg)
    if deg==1:
        x=data.iloc[:,:-1].values
        y=data.iloc[:,1].values
        Model=LinearRegression()
        Model.fit(x,y)
        
    else:
        x=feats
        y=data['y']
        Model=Ridge(alpha)
        Model.fit(x,y)
    return Model
```

## Perform a Ridge Fit of a 16th Degree Polynomial Using a Very Small Penalty Strength

```python
model=poly_ridge_reg(data, deg=16, alpha=1e-25)
PCoeff(model)
PPP(data, model)
```

## Perform a Ridge fit of a 16th Degree Polynomial Using a Very Large Penalty Strength

```python
model=poly_ridge_reg(data,deg=16,alpha=100)
PCoeff(model)
PPP(data, model)
```

## Let's look at fits for a sequene of increasing lambda values

```python
for alpha in [1e-25,1e-10,1e-6,1e-3,1,1e2]:
    model=poly_ridge_reg(data,deg=16, alpha=alpha)
    print("Alpha: "+str(alpha)+'\n')
    PCoeff(model)
    plt.figure()
    PPP(data,model)
    plt.title('Ridge, Alpha: '+str(alpha)+'\n')
```

## Perform a ridge fit of a 16th degree polynomial using a "good" penalty strength

We will learn about cross validation in the course as a way to select a good value of the turning parameter lambda/alpha. Here, we consider "Leave one out" (LOO) cross validation, which one can show approximates average mean square erros (MSE). As a result, choosing lambda/alpha to minimize the LOO error is equivalent to choosing lambda/alpha to minimize an approximation to average MSE.

```python
from sklearn.model_selection import KFold
```

```python
def LOO(data,deg,alpha_val):
    #Define Feats and "x,y" values
    feats=poly_fts(data,deg)
    if deg==1:
        x=data.iloc[:,:-1].values
        y=data.iloc[:,1].values
        
    else:
        x=feats
        y=data['y']
    
    #Split data indices
    num_folds=len(data)
    kf=KFold(num_folds)
    
    alpha_mse=[]
    min_mse=None
    best_alpha=None
    for alpha in alpha_val:
        next_mse=0
        for train_index,valid_index in kf.split(data):
            train_set_x=pd.DataFrame(columns=x.columns)
            train_set_y=pd.DataFrame(columns=['y'])
            for train_val in train_index:
                val={'y':[y[train_val]]}
                train_df_x=x.iloc[[train_val]]
                train_df_y=pd.DataFrame(val)
                train_set_x=pd.concat([train_set_x,train_df_x],ignore_index=True)
                train_set_y=pd.concat([train_set_y,train_df_y],ignore_index=True)
            
            valid_set_x=pd.DataFrame(columns=x.columns)
            valid_set_y=pd.DataFrame(columns=['y'])
            for valid_val in valid_index:
                val={'y':[y[valid_val]]}
                valid_df_x=x.iloc[[valid_val]]
                valid_set_x=pd.concat([valid_set_x,valid_df_x],ignore_index=True)
                valid_df_y=pd.DataFrame(val)
                valid_set_y=pd.concat([valid_set_y,valid_df_y],ignore_index=True)
        
            
            model=Ridge(alpha)
            model.fit(train_set_x,train_set_y)
            
            y_test_pred=model.predict(valid_set_x)
            next_mse+=((y_test_pred-valid_set_y)**2).sum()
            
        next_mse=next_mse/num_folds
        alpha_mse.append(float(next_mse))
        if min_mse is None or float(next_mse)<float(min_mse):
            min_mse=next_mse
            best_alpha=alpha
            
    return alpha_mse, best_alpha
```

```python
alpha_val=np.logspace(-4,10,num=10)
alpha_mse,best_alpha=LOO(data,16,alpha_val)
```

```python
print(alpha_mse)
print()
print(best_alpha)
```

```python
plt.plot(alpha_val,alpha_mse,'k-')
plt.xlabel('$\L2_penalty$')
plt.ylabel('LOO cross validation error')
plt.xscale('log')
plt.yscale('log')
```

```python
model=Ridge(best_alpha)
model.fit(poly_fts(data,16),data['y'])
PCoeff(model)
PPP(data, model)
```
